# Multiple answer questions

```python
def ExtractMAQ ( x ):
    data = x
    #---------------------------------------------------------    
    # Extract the target column
    data = data[data.columns[0]].str.split(';', expand=True)     
    #---------------------------------------------------------    
    #Creat a unique list from the columns
    var_data = []

    for x in data.columns:
        for tab in data.index:
            var_data.append(data[x][tab])
    var_df = pd.DataFrame (var_data).rename(columns = {0:"var"}) # Convert to dataframe 
    var_df = var_df [var_df['var'].str.contains ('Nan') == False]
    var_list = var_df['var'].unique().tolist()
    #var_list    
    #---------------------------------------------------------    
    # creat columns named as the list
    for x in var_list : 
        data[x] = 0    
    #---------------------------------------------------------
    for col in data:  # Replace the nan value to #
        if isinstance(col, str) == False :  
            data[col].replace (np.nan , "#" , inplace = True)    
    #---------------------------------------------------------
    for col in data: # filling the columns we have created by
        if isinstance(col, str) == False : 
            for row in data.index:
                value = data[col][row]
                data.loc[row , value ] = 1   
                
#-----------------Drop the columns ----------------------------------------
    data2 = data.drop("#" , axis = 1)      
    for col in data : 
        if isinstance(col, str) == False :
            data2 = data2.drop(col , axis = 1 ) 
            
#----------------reset the index--------------------------------------------        
    data2.reset_index(inplace = True )    

#--------------- pivot the columns------------------------------------------    
    res = pd.melt(data2, id_vars=['index'], var_name=['Type'] )
    res = res[res['value'] != 0]
    res.drop ('value' , axis = 1 , inplace = True)
    return(res)    
```

```python
ColumnsShape = []

for i in MultipleChoices:
    dd = ExtractMAQ (df[[i]])
    print ('----{}---'.format (i))
    ColumnsShape.append ({'CoulmnsName':i , 'Shape':dd.shape[0]})
    #print ('The shape is:{}'.format (dd.shape[0]))
    #display (dd.head(5))
    dd.to_excel ('.\\Data\\{}_DF.xlsx'.format (i))
```

# Salary

```python
# Outlier function
def outlier_treatment(datacolumn):
    sorted(datacolumn)
    Q1 , Q3 = np.percentile(datacolumn , [25,75])
    IQR = Q3 - Q1
    upper_range = Q3 + (1.5 * IQR)
    return upper_range
```

```python
def salaryf (data):
    # 1.Create Dataframe for columns that related to salary
    SalaryDF = data[salary]
    SalaryDF = SalaryDF[SalaryDF['CompFreq'].str.contains ('NaN') == False]  
    
    # 2.Remove the non values
    SalaryDF = SalaryDF.dropna()
    
    # 3.Extract currency type from Currency column.
    SalaryDF['Currency_Type'] = SalaryDF['Currency'].apply(lambda x: x[0:3])
    SalaryDF['Currency_Describtion'] = SalaryDF['Currency'].apply(lambda x : x[4:])
    #----------------------------------------------------------------------------#
    #Drop the Currency column
    SalaryDF.drop('Currency' , axis = 1 , inplace = True)
    #----------------------------------------------------------------------------#
    
    SalaryDF = SalaryDF[SalaryDF['CompTotal']< 5000000]
    
    # 5.Convert the salaries to annual salaries
    SalaryDF['TotalByYear'] = 0

    # fill the total salary by year
    for x in SalaryDF.index:
        condition = SalaryDF.loc[x , 'CompFreq']
        if condition == 'Yearly' :
            SalaryDF.loc[x ,'TotalByYear'] = SalaryDF.loc[x , 'CompTotal'] * 1

        if condition == 'Monthly':
            SalaryDF.loc[x ,'TotalByYear'] = SalaryDF.loc[x , 'CompTotal'] * 12

        if condition == 'Weekly':
            SalaryDF.loc[x ,'TotalByYear'] = SalaryDF.loc[x , 'CompTotal'] * 52

            
    # 6.Collect the latest Exchange rates.
            
    url = 'https://www.xe.com/currencytables/?from=USD'
    dfs = pd.read_html(url)
    rate = pd.DataFrame (dfs[0]).set_index('Currency')
    
    
    
    # 7.Convert all salaries to USD
            
    # add columns for total USD salary by year
    SalaryDF ['TotalByYearUSD'] = 0
    SalaryDF['Echange_rate'] = 0
    

        # Searching to currency type and calculate the annual salary by exchange rate
    for x in SalaryDF.index :
        try:
            currency = SalaryDF.loc[x , 'Currency_Type']
            exchange = rate.loc[ currency , 'USD per unit']
            SalaryDF.loc[ x , 'TotalByYearUSD' ] = SalaryDF.loc[x , 'TotalByYear'] * exchange
            SalaryDF.loc[x ,'Echange_rate'] = exchange 
            print ("Currency:" , currency , "-----" , "Exchage:" , exchange)
        except:
            print ("Expect" ,"Currency:" , currency , "-----" , "Exchage:" , exchange)
            #print (currency)

    # Filter the saly above of 10$ and lowerthan outlier
    SalaryDF = SalaryDF[ (SalaryDF['TotalByYearUSD'] > 10) & 
                    (SalaryDF['TotalByYearUSD'] < outlier_treatment(SalaryDF.TotalByYearUSD)) ]
    
    return SalaryDF
```

```python
salary = salaryf (df)
```

```python
# 8.Export the final data to EXCEL.
salary.reset_index().to_excel ('.\\Data\\Salary_DF.xlsx')
```
